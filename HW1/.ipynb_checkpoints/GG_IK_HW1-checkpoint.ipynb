```python
# import packages
import pandas as pd
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
from scipy.linalg import hadamard
import sympy 
import time
from numba import jit
```

```python
# 4. Randomized matrix multiplication:

#  A. Read in the matrices 
A = np.matrix(pd.read_csv('STA243_homework_1_matrix_A.csv', header=None))
B = np.matrix(pd.read_csv('STA243_homework_1_matrix_B.csv', header=None))

#  B. Calculate the product
M = A @ B

#  C. Algorithm 2: Random Matrix Multiplication Algorithm
def alg2(A, B, r):
    """
    Perform matrix multiplication using Random Matrix Multiplication algorthim. 
    
    Use r outer products to estimate the prodcut of two matrices A and B. 

    Parameters
    ----------
    A : np.ndarray(m, n)
        Matrix A with m rows and n columns stored in a numpy ndarray 
    B : np.ndarray(n, p)
        Matrix B with n rows and p columns stored in a numpy ndarray 
    r : int
        number of 

    Returns
    -------
    M : np.ndarray(m, p)
        Product of A and B stored in a numpy ndarray
    """

    # get summary stats
    n = A.shape[1]

    # calculate p_k
    num = [norm(A[:,k], 2) * norm(B[k,:], 2) for k in range(n)]
    p_k = num/sum(num)
    
    # sample r random samples from 1 through n with prob p_k
    x_l = np.random.choice(a = range(n), size = r, replace=True, p=p_k)
    
    # subset to just randomly picked x_l (outer products)
    M = sum([np.outer(A[:,i], B[i,:])/(p_k[i] * r) for i in x_l])

    return M


#  D. Function to calculate errors
def calc_error(M, A, B):
    # calculate relative approximation error using frobenius norm
    error = norm(M - A@B, 'fro') / (norm(A, 'fro') * norm(B, 'fro'))

    return error


#  E. Calculate relative error
rel_error = {r: calc_error(alg2(A, B, r), A, B) for r in [20, 50, 100, 200]}
print(rel_error)

#  F. Visualize approximations 
approx = [alg2(A, B, r) for r in [20, 50, 100, 200]]
for i in range(len(approx)):
    plt.figure(figsize=(7,5))
    plt.imshow(approx[i])
    plt.show()
```

```python
# 5. Power method
#  A. Define power iteration 
def power_iteration(A, v0, eps = 1e-6, maxiter=100):
    """
    Estimate largest eigenvector using Power Method algorithm.

        Parameters
    ----------
    A : np.ndarray(n, n)
        Matrix A with m rows and n columns stored in a numpy ndarray 
    v0: np.array (n,)
        Matrix B with n rows and p columns stored in a numpy ndarray 

    Returns
    -------
    v : np.array(n, )
        Product of A and B stored in a numpy ndarray

    """
    # initialize algorithm
    t = 1
    v_prev = v0
    diff = np.inf

    # run algorithm until 
    #  (1) max iterations is reached or 
    #  (2) norm of the differences between iteration is less than threshold
    while t < maxiter and diff > eps:
        # find new vector and normalize 
        v_next = A @ v_prev
        v_next = v_next / norm(v_next, 2)

        # find the norm of differences between the iterations 
        diff = norm(v_next - v_prev)

        # advance iteration counter
        t = t + 1

        # update the v_prev
        v_prev = v_next
    
    return(v_next)

#  B. Compare estiamtes 
np.random.seed(2022)
E = np.random.normal(size=(10, 10))
v = np.array([1] + [0]*9)
lams = np.arange(1, 11)
prods = []
for lam in lams:
    X = lam*np.outer(v, v) + E
    v0 = np.ones(10)
    v0 = v0/np.linalg.norm(v0, 2)
    vv = power_iteration(X, v0)
    prods.append(np.abs(v @ vv))

# plot results 
plt.plot(lams, prods, '-ok')
plt.xlabel('lambda')
plt.ylabel('product')
plt.show()
```

```python
# 6. Sketching for Least-squares:

#  A. Define helper functions 
#   i. Fast Walsh-Hadamard Transform
@jit(nopython=True)
def fwht(a) -> None:
    """In-place Fast Walsh-Hadamard Transform of array a."""
    h = 1
    while h < len(a):
        for i in range(0, len(a), h * 2):
            for j in range(i, i + h):
                x = a[j]
                y = a[j + h]
                a[j] = x + y
                a[j + h] = x - y
        h *= 2

#   ii. Flatten array 
def flatten(listOfLists):
    "Flatten nesting"
    return np.array([x[0] for x in listOfLists])

#   iii. Phi transformation 
def phi_tr(X, D_tr, t, r):
    """
    Perform subsampled randomized linear tranformation. Phi = S^tHD
    
    X: np.array(n, d)
    D_tr: np.array(n, ) n length vector of random sample from {-1, 1} 
    t: indices of the subsampled rows
    r: number of subsampled rows 
    """
    # Apply transformation one-after-another
    # 1. Apply D 
    DX = D_tr * X

    # 2. Apply H (use the in-place fast walsh alg)
    for i in range(DX.shape[1]):
        fwht(DX[:, i])
    
    # 3. Apply S^t
    phi_X = DX[t] / np.sqrt(r)

    return(phi_X)


#   iv. Compare ols to sketched ols using l2 norm
def comp_to_ols(X, Y, b_s):
    """
    Compare sketched OLS beta coefficients to OLS solution
    b_s: X: np.array (d, ) vector of sketched OLS coefficients
    X: np.array (n, d)
    Y: np.array (n, )
    """

    # A. Calculate the OLS solution 
    ols = np.linalg.inv(X.T @ X) @ X.T @ Y

    # B. Compare 
    diff = np.linalg.norm(flatten(ols) - flatten(b_s), 2)

    return(diff)

#   v. Run the sketched ols algorithm n times 
def run_comp_n_times(X, Y, eps, n):
    """
    Run sketched OLS on the same X, Y n times and return the average error
    """

    # A. Run ols solution n time, save each result to a list
    b_s = [sketched_ols(X, Y, eps) for x in range(n)]
    
    # split the result into times and coefficients
    coef = [b_s[i][0] for i in range(len(b_s))]
    tms = [b_s[i][0][1] for i in range(len(b_s))]

    # B. Calculate error
    mean_coef_error = np.mean([comp_to_ols(X, Y, coef[i]) for i in range(len(coef))])
    
    return(mean_coef_error, tms)


#  B. Define sketched ols algorithm 
def sketched_ols(X, y, eps, talk_flg = True):
    """
    X: np.array (n, d)
    y: np.array (n, )
    epsilon: constant (0, 1)
    """

    # initialize variables 
    n = X.shape[0]
    d = X.shape[1] 
    r = int(np.round((d * np.log(n))/eps))

    # sampling 
    D_tr = np.random.choice(a=[-1, 1], size=n, replace=True).reshape(n, 1)
    t = np.random.choice(a=range(n), size=r, replace=True)
   
    # calc x_star
    x_star = phi_tr(X, D_tr, t, r).astype('float64')

    # calc y_star
    y_star = phi_tr(y, D_tr, t, r).astype('float64')

    ts1 = time.time()
    b_s = np.linalg.inv(x_star.T @ x_star) @ x_star.T @ y_star
    ts2 = time.time()
    tm = ts2 - ts1
    if (talk_flg == True ):
        print("Last calc for epsilon {} was done in {} seconds".format(eps, round(tm, 5)))
   
    return(b_s, tm)



#  C. Generate a 1048576 × 20 design matrix X and a 1048576 × 1 response y with elements drawn iid from a Uniform(0, 1) distribution
np.random.seed(1)
X = np.random.uniform(size=(1048576, 20))
Y = np.random.uniform(size=(1048576, 1))

#  D. Record ols calculation time
ts_start = time.time()
ols = np.linalg.inv(X.T @ X) @ X.T @ Y
ts_end = time.time()
print("OLS calculation time: {}".format(round(ts_end - ts_start, 5)))

#  E. Compare calcuations times for a single estimate per epsilon 
eps_list = [0.1, 0.05, 0.01, 0.001]
chk = [run_comp_n_times(X, Y, eps, 1) for eps in eps_list]
```
