### a)

```python
# A is m x n, B is n x p
def rand_matrix_mult(A, B, r):
    if A.shape[1] != B.shape[0]:
        print('The number of columns in Matrix A needs to be equal to the number of rows in Matrix B')
    
    # define n and initialize M as empty matrix
    n = A.shape[1]
    M = np.zeros([A.shape[0], B.shape[1]])
    
    # define probabilities
    norms = [norm(A[:,k], ord=2) * norm(B[k,:], ord=2) for k in range(n)]
    p_k = norms / np.sum(norms)
    
    i_l = np.random.choice(a=np.arange(n), replace=True, size=r, p=p_k)
    
    for i in i_l:
        M += 1/(r * p_k[i]) * np.outer(A[:, i], B[i, :])
    
    return np.round(M, 6)
```

### b)

```python
M_r20 = rand_matrix_mult(matrixA, matrixB, 20)
M_r50 = rand_matrix_mult(matrixA, matrixB, 50)
M_r100 = rand_matrix_mult(matrixA, matrixB, 100)
M_r200 = rand_matrix_mult(matrixA, matrixB, 200)
```

### c)

```python
def relative_approx_error(M, A, B):
    error = np.linalg.norm(M - A @ B, ord='fro') / (np.linalg.norm(A, ord='fro') * np.linalg.norm(B, ord='fro'))
    return error

r_list = [20, 50, 100, 200]
errors = [relative_approx_error(rand_matrix_mult(matrixA, matrixB, r), matrixA, matrixB) for r in r_list]
```

```python
plt.figure(figsize=(7,5))
plt.plot(r_list, errors)
plt.scatter(r_list, errors)
plt.xlabel('r')
plt.ylabel('relative approximation error')
plt.title('relative approximation error vs r')
plt.savefig('rand_linalg_plot.png')
plt.show()
```

```python
def power_iteration(A, v0, eps = 1e-6, maxiter=100):
    """
    Please implement the function power_iteration that takes in the matrix X and initial vector v0 and 
    returns the eigenvector.
    A: np.array (d, d)
    v0: np.array (d,)
    """
    for i in range(maxiter):
        c = A @ v0
        est_eigenvector = c/np.linalg.norm(c)
        error = np.linalg.norm(v0 - est_eigenvector)
        if error < eps:
            return est_eigenvector
        v0 = est_eigenvector
        

if __name__ == '__main__':

    np.random.seed(2022)
    E = np.random.normal(size=(10,10))
    v = np.array([1]+[0]*9)
    lams = np.arange(1, 11)
    prods = []
    for lam in lams:
        X = lam*np.outer(v,v) + E
        v0 = np.ones(10)
        v0 = v0/np.linalg.norm(v0,2)
        vv = power_iteration(X, v0)
        prods.append(np.abs(v @ vv))

    plt.plot(lams, prods, '-ok')
    plt.xlabel('lambda')
    plt.ylabel('product')
    plt.savefig('powerMethod_plot.png')
    plt.show()
```

```python
def sketched_OLS(X, y, eps):
    n = X.shape[0]
    d = X.shape[1]
    r = ((d * np.log(n))/eps).astype('int')
    S = np.empty(shape=[n,r]) # n x r
    D = np.identity(n) # n x n
    for i in range(r):
        rand_num = np.random.choice(np.arange(1, n+1), size=1, replace=True) # pick random number from 1 to n
        if i == rand_num:
            e = np.zeros(n) # initialize e
            e[i] = 1 # set ith position equal to 1
            S = np.column_stack((S, np.sqrt(n/r) * e)) # append to S
    
    D = np.diag(np.random.choice([-1,1], size=n, p=[0.5,0.5]))
    H = (1/np.sqrt(n)) * hadamard(n) # n x n
    phi = S.T @ H @ D
    beta = inv((phi @ X).T @ phi @ X) @ (phi @ X).T @ phi @ y
    return beta
```
